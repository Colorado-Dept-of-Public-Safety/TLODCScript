```python
#Step 1: Import Libraries and Data
# Libraries like pandas and numpy can handle almost al of the data cleaning we will do for this dataset. 
import pandas as pd 
import numpy as np 
import os # you'll need this library for exporting out the cleaed out CSV. 
import dataprep
import regex
from dataprep.clean import clean_phone

#  import data consists of three steps: 
file = r'C:/Users/rstevenson/Desktop/Desktop Folder/CDPS/DHSEM/CIAC/TLO/TLO_23_24.xlsx' # Brings in a clean copy of the data 
TLO_Data = pd.read_excel(file) # Creates a Data Frame 
TLO_Data.head() # Outputs the dataframe's first 5 entries.
```

#### Step 1: Clean Up Address1 Column ####
<p> The Address1 column needs to be cleaned so that it is proper case and that Ste are removed. </p>

```python
# Remove keyword "Ste" and anything that follows it from Address1 column. 
TLO_Data['Address1'] = TLO_Data['Address1'].str.replace(r'(?i)\s*\b(Ste|suite|Suite|Rm|Room|Building|#|BLDG).*', '')
pd.set_option('display.max_rows', 510)
TLO_Data.head()
```

```python
# Setup proper case for all addresses 
TLO_Data["Address1"]= TLO_Data["Address1"].str.upper().str.title()
TLO_Data.tail()
```

#### Clean Up First and Last Name Columns and PositionTitleRank, AgencyOrganization and City ####

```python
TLO_Data["FirstName"] = TLO_Data["FirstName"].str.upper().str.title()
TLO_Data["LastName"] = TLO_Data["LastName"].str.upper().str.title()
TLO_Data["PositionTitleRank"] = TLO_Data["PositionTitleRank"].str.upper().str.title()
TLO_Data["AgencyOrganization"] = TLO_Data["AgencyOrganization"].str.upper().str.title()
TLO_Data
```

<h2> Clean Up Additional Columns </h2>


#### Clean Up Phone Column ####
